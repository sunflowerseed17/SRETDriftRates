# Cleaning Quickly


```python
# Loading in the dependencies

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import IPython.display as display
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.table as tbl
import textwrap

```

```python

# Load data
drift_rates = pd.read_csv('pyddm_drift_alpha_bias.csv')

# Fix typo
drift_rates['Category'] = drift_rates['Category'].replace({'low domonance': 'low dominance'})

# Normalize spacing and case
drift_rates['Category'] = drift_rates['Category'].str.strip().str.lower()

# Map each full category label to (Dimension, Simplified Category)
category_map = {
    'positive': ('Valence', 'Positive'),
    'negative': ('Valence', 'Negative'),
    'low affiliation': ('Affiliation', 'Low'),
    'high affiliation': ('Affiliation', 'High'),
    'low dominance': ('Dominance', 'Low'),
    'high dominance': ('Dominance', 'High'),
}

# Apply mapping
drift_rates['Dimension'] = drift_rates['Category'].map(lambda x: category_map.get(x, ('Unknown', ''))[0])
drift_rates['Category'] = drift_rates['Category'].map(lambda x: category_map.get(x, ('', ''))[1])

# Build Dimension_Category
drift_rates["Dimension_Category"] = drift_rates["Dimension"] + " | " + drift_rates["Category"]
```

# Drift Rates



```python


# Load data
df = pd.read_csv('pyddm_drift_alpha_bias.csv')

# Fix typo and normalize case
df['Category'] = df['Category'].replace({'low domonance': 'low dominance'})
df['Category'] = df['Category'].str.strip().str.title()

# Define color palette
palette = {
    "Negative": "#FF8F82",
    "Positive": "#6DFFAAB0",
    "Low Affiliation": "#FF8F82",
    "High Affiliation": "#6DFFAAB0",
    "Low Dominance": "#FF8F82",
    "High Dominance": "#6DFFAAB0",
}

# Filter for only the selected categories
selected_categories = list(palette.keys())
filtered_df = df[df["Category"].isin(selected_categories)]

# Loop over each category and plot individually
for category in selected_categories:
    subset = filtered_df[filtered_df["Category"] == category]
    color = palette.get(category, "gray")

    plt.figure(figsize=(8, 5))
    sns.histplot(
        data=subset,
        x="DriftRate",
        kde=True,
        stat="density",
        bins=30,
        color=color,
        alpha=0.6
    )

    plt.title(f"Drift Rate Distribution: {category}", fontsize=16)
    plt.xlabel("Drift Rate", fontsize=12)
    plt.ylabel("Density", fontsize=12)
    plt.axvline(0, linestyle='--', color='gray')
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

# Regression


## Drift Rate


```python
print(df.columns.tolist())
conditions = sorted(df['Category'].unique())
print(f"Unique conditions found: {conditions}\n")

# 3. Separate BDI and LSAS data for ease of analysis
df_bdi = df[df['Measure'] == 'BDI'].copy()
df_lsas = df[df['Measure'] == 'LSAS'].copy()

# 4-6. Correlation, Regression, and Effect Size for each condition
for cond in conditions:
    # Filter data for the current condition
    sub_bdi = df_bdi[df_bdi['Category'] == cond].copy()
    sub_lsas = df_lsas[df_lsas['Category'] == cond].copy()
    # Drop any rows with missing DriftRate or Score
    sub_bdi.dropna(subset=['DriftRate', 'Score'], inplace=True)
    sub_lsas.dropna(subset=['DriftRate', 'Score'], inplace=True)
    
    # Skip condition if no data (or only one participant) is available
    if len(sub_bdi) < 2 or len(sub_lsas) < 2:
        print(f"Condition: {cond}")
        print("  [Not enough data to compute statistics]\n")
        continue
    
    # Extract arrays for DriftRate and scores
    drift_bdi = sub_bdi['DriftRate'].values
    scores_bdi = sub_bdi['Score'].values  # BDI scores
    drift_lsas = sub_lsas['DriftRate'].values
    scores_lsas = sub_lsas['Score'].values  # LSAS scores
    
    # Pearson correlations
    r_bdi, p_bdi = pearsonr(drift_bdi, scores_bdi)
    r_lsas, p_lsas = pearsonr(drift_lsas, scores_lsas)
    
    # Linear regressions (DriftRate ~ Score)
    # Reshape scores to 2D for sklearn
    X_bdi = scores_bdi.reshape(-1, 1)
    X_lsas = scores_lsas.reshape(-1, 1)
    linreg_bdi = LinearRegression().fit(X_bdi, drift_bdi)
    linreg_lsas = LinearRegression().fit(X_lsas, drift_lsas)
    # Regression parameters
    slope_bdi = linreg_bdi.coef_[0]
    intercept_bdi = linreg_bdi.intercept_
    r2_bdi = linreg_bdi.score(X_bdi, drift_bdi)  # R^2 for BDI model
    slope_lsas = linreg_lsas.coef_[0]
    intercept_lsas = linreg_lsas.intercept_
    r2_lsas = linreg_lsas.score(X_lsas, drift_lsas)  # R^2 for LSAS model
    
    # Cohen's f^2 effect size for each regression model
    # (If R^2 is 1 or very close, handle to avoid division by zero)
    f2_bdi = np.inf if r2_bdi >= 1 else r2_bdi / (1 - r2_bdi)
    f2_lsas = np.inf if r2_lsas >= 1 else r2_lsas / (1 - r2_lsas)
    
    # 5. Print the results for this condition
    print(f"Condition: {cond}")
    # BDI results
    print(f"  Pearson correlation (DriftRate vs BDI): r = {r_bdi:.3f}, p-value = {p_bdi:.3g}")
    print(f"  Linear regression (DriftRate ~ BDI): slope = {slope_bdi:.3f}, intercept = {intercept_bdi:.3f}, R^2 = {r2_bdi:.3f}, Cohen's f^2 = {f2_bdi:.3f}")
    # LSAS results
    print(f"  Pearson correlation (DriftRate vs LSAS): r = {r_lsas:.3f}, p-value = {p_lsas:.3g}")
    print(f"  Linear regression (DriftRate ~ LSAS): slope = {slope_lsas:.3f}, intercept = {intercept_lsas:.3f}, R^2 = {r2_lsas:.3f}, Cohen's f^2 = {f2_lsas:.3f}\n")
    
    # 6. Scatterplots with regression lines for this condition
    # Plot DriftRate vs BDI
    plt.figure(figsize=(5, 4))
    plt.scatter(scores_bdi, drift_bdi, color="#5353F5", alpha=0.7, label='Data (BDI)')
    # Regression line for BDI
    line_x = np.linspace(min(scores_bdi), max(scores_bdi), 100)
    line_y = slope_bdi * line_x + intercept_bdi
    plt.plot(line_x, line_y, color="#5353F5", linewidth=2, label=f'Fit: Drift ~ BDI')
    plt.xlabel('BDI Score')
    plt.ylabel('Drift Rate')
    plt.title(f'DriftRate vs BDI ({cond})')
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    # Plot DriftRate vs LSAS
    plt.figure(figsize=(5, 4))
    plt.scatter(scores_lsas, drift_lsas, color="#43C6B690", alpha=0.7, label='Data (LSAS)')
    # Regression line for LSAS
    line_x = np.linspace(min(scores_lsas), max(scores_lsas), 100)
    line_y = slope_lsas * line_x + intercept_lsas
    plt.plot(line_x, line_y, color="#43C6B690", linewidth=2, label=f'Fit: Drift ~ LSAS')
    plt.xlabel('LSAS Score')
    plt.ylabel('Drift Rate')
    plt.title(f'DriftRate vs LSAS ({cond})')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
